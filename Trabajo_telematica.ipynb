```python
# Solo copia y pega esta celda en tu notebook y ejecútala.
# Para modificar los datos, edita directamente el diccionario `data`.

import pandas as pd
import matplotlib.pyplot as plt

# ————————————————————————
# 1) DEFINE TUS DATOS (modifica aquí si quieres)
# ————————————————————————
data = {
    'Nombre': [
        'GUTIERREZ MEZA RONALD ORLANDO',
        'LOPEZ RODRIGUEZ MARIO ALEJANDRO',
        'RODRIGUEZ TAPIAS RONALD STEVEN',
        'MURCIA CALDERON YÉSICA VIVIANA',
        'RAMIREZ LOPEZ INGRID JULIETH',
        'SIERRA VELANDIA PABLO ENRIQUE',
        'NIETO ARANGUREN JOSÉ GREGORIO',
        'BERON GUERRA RENSO',
        'LAYOS TABARES CHRISTIAN DAVID',
        'SILVA CUCHUMBE JOSE ANTONIO',
        'RINCÓN SARMIENTO JOSE CESAR',
        'HOYOS HOYOS EIMER DAVID',
        'GÓMEZ GALVIS ANGIE',
        'HENAO QUITIAN WRAYAN FELIPE'
    ],
    'Edad':    [28, 24, 24, 22, 21, 26, 40, 39, 37, 41, 29, 34, 21, 21],
    'Estatura':[1.68, 1.65, 1.72, 1.50, 1.50, 1.67, 1.65, 1.72, 1.70, 1.65, 1.65, 1.65, 1.64, 1.67],
    'Peso':    [60, 65, 71, 58, 59, 68, 63, 70, 71, 80, 65, 72, 62, 58]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Función auxiliar para graficar
def plot_barras(columna, color, ylabel, title):
    plt.figure(figsize=(12, 6))
    plt.bar(df['Nombre'], df[columna], color=color)
    plt.xticks(rotation=90, ha='right')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.tight_layout()
    plt.show()

# ————————————————————————
# 2) GRAFICA CADA VARIABLE
# ————————————————————————
plot_barras('Edad',    'skyblue',    'Edad (años)',         'Edad de los estudiantes')
plot_barras('Estatura','lightgreen', 'Estatura (m)',        'Estatura de los estudiantes')
plot_barras('Peso',    'salmon',     'Peso (kg)',           'Peso de los estudiantes')
```
