```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de 10 alumnos
alumnos = [f'Alumno {i+1}' for i in range(10)]
edades  = [18, 19, 17, 20, 18, 19, 17, 18, 20, 19]
alturas = [1.65, 1.70, 1.60, 1.75, 1.68, 1.72, 1.58, 1.64, 1.78, 1.69]
pesos   = [60,   65,   55,   70,   62,   68,   54,   61,   72,   64]

# Posiciones y ancho de barra
x     = np.arange(len(alumnos))
ancho = 0.25

# Crear gr√°fico
plt.figure(figsize=(12,6))
plt.bar(x - ancho, edades,  ancho, label='Edad')
plt.bar(x,         alturas, ancho, label='Altura')
plt.bar(x + ancho, pesos,   ancho, label='Peso')

plt.xlabel('Alumnos')
plt.ylabel('Valores')
plt.title('Edad, Altura y Peso de 10 Alumnos')
plt.xticks(x, alumnos, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
